**2. Use Crime dataset from LMS**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path="https://github.com/RudreshKumarP/Python-ExcelR/raw/main/crime_data.csv"
sample=pd.read_csv(path)
```

```python
sample.head()
```

```python
sample.rename( columns ={'Unnamed: 0' : 'Cities'}, inplace=True)
sample.head()
```

```python
sample['UrbanPop'].count()
```

```python
sample.sum()
```

```python
sample.agg(['median', 'mean', 'max', 'min', 'std'])
```

**II) do the plottings like plottings like histogram, boxplot, scatterplot, barplot, piechart,dot chart.**

```python
plt.hist(sample['UrbanPop'])
plt.show()
```

```python
plt.figure(figsize = (12,12))
plt.scatter (sample.Murder, sample.Cities)
plt.title ('Cities Vs Murder')
plt.xlabel ('Murder')
plt.ylabel ('Cities')
plt.grid()
plt.show()
```

```python
plt.boxplot(sample.Assault)
plt.show()
```

```python
State = sample.head().pivot_table(columns=['State'])
print(State)
```

```python
State = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California']
Assault = [236, 263, 294, 190, 276]
plt.pie(Assault, labels=State, autopct='%.0f%%')
plt.title('Total Number of Assaults in each State')
plt.show()
```

```python
plt.figure (figsize= (8,8))
plt.xlabel('Number')
plt.ylabel('Population')
sample.head()['UrbanPop'].plot(kind='bar')
plt.show()
```
